```python
import os

# Chemin du dossier GTFS
gtfs_dir = r"C:\Users\tiend\Downloads\Toulouse_GTFS"  # <-- adapte selon la ville

def detect_separator(file_path):
    # Essaye de détecter le séparateur en lisant la première ligne
    with open(file_path, 'r', encoding='utf-8') as f:
        sample = f.readline()
        for sep in [',', ';', '\t']:
            if sep in sample:
                return sep
    return ','  # Par défaut

def count_lines_without_header(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        # S'il y a au moins une ligne (le header), on ne le compte pas
        if len(lines) <= 1:
            return 0
        return len(lines) - 1

print("Fichier\t\tNombre de lignes\tSéparateur")
print("-" * 60)

for filename in sorted(os.listdir(gtfs_dir)):
    if filename.endswith(".txt"):
        file_path = os.path.join(gtfs_dir, filename)
        sep = detect_separator(file_path)
        n_lines = count_lines_without_header(file_path)
        if sep == ',':
            sep_str = 'virgule (,)'
        elif sep == ';':
            sep_str = 'point-virgule (;)'
        elif sep == '\t':
            sep_str = 'tabulation'
        else:
            sep_str = repr(sep)
        print(f"{filename:20s}\t{n_lines:5d}\t\t{sep_str}")
```
